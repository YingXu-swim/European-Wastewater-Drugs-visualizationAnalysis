# 整理成国家数据

```python
DAILY = dict()
for i in range(len(data)):
    country = data.country[i]
    if country in DAILY:
        if DAILY[country][data.year[i]][data["metabolite"][i]] == None:
            DAILY[country][data.year[i]][data["metabolite"][i]] = data["Daily mean"][i]
        else:
            DAILY[country][data.year[i]][data["metabolite"][i]] += data["Daily mean"][i]
    else:
        DAILY[country] = {i:{j:None for j in set(data.metabolite)} for i in range(2011,2022)}
```

```python
WEEKDAY = dict()
for i in range(len(data)):
    country = data.country[i]
    if country in WEEKDAY:
        if WEEKDAY[country][data.year[i]][data["metabolite"][i]] == None:
            WEEKDAY[country][data.year[i]][data["metabolite"][i]] = data["Weekday mean"][i]
        else:
            WEEKDAY[country][data.year[i]][data["metabolite"][i]] += data["Weekday mean"][i]
    else:
        WEEKDAY[country] = {i:{j:None for j in set(data.metabolite)} for i in range(2011,2022)}
```

```python
WEEKEND = dict()
for i in range(len(data)):
    country = data.country[i]
    if country in WEEKEND:
        if WEEKEND[country][data.year[i]][data["metabolite"][i]] == None:
            WEEKEND[country][data.year[i]][data["metabolite"][i]] = data["Weekend mean"][i]
        else:
            WEEKEND[country][data.year[i]][data["metabolite"][i]] += data["Weekend mean"][i]
    else:
        WEEKEND[country] = {i:{j:None for j in set(data.metabolite)} for i in range(2011,2022)}
```

```python
DAILY['CH']
```

```python
country_ls = []
year_ls = []
daily = []
weekday = []
weekend = []
medcine = []

for country in DAILY:
    for year in DAILY[country]:
        for med in DAILY[country][year]:
            country_ls.append(country)
            year_ls.append(year)
            daily.append(DAILY[country][year][med])
            weekday.append(WEEKDAY[country][year][med])
            weekend.append(WEEKEND[country][year][med])
            medcine.append(med)
```

```python
raw = pd.DataFrame({"year":year_ls,"counry":country_ls,"metabolite":medcine,
                    "daily mean":daily,"weekday mean":weekday,"weekend mean":weekend})
raw
```
